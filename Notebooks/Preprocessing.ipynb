```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
```

```python
raw_data = pd.read_csv('pid_dataset_with_ss_error.csv')
data = raw_data.copy()
data
```

```python
from sklearn.utils import shuffle
data = shuffle(data, random_state=45)
data.columns.values
```

```python
column_to_scale =  ['RiseTime', 'Overshoot', 'SettlingTime','SS_Error', 'Kp', 'Ki', 'Kd']

scaler = StandardScaler()
scaled = scaler.fit_transform(data[column_to_scale])
data_scaled = pd.DataFrame(scaled, columns=column_to_scale)
data_scaled
```

```python
data_input = data_scaled[['RiseTime', 'Overshoot', 'SettlingTime','SS_Error']]
data_target = data_scaled[['Kp', 'Ki', 'Kd']]
```

```python
n = len(data_scaled)
train_size = int(n*0.8)
validation_size = int(n*0.1)
test_size = n - train_size - validation_size

train_input = data_input[:train_size]
validation_input = data_input[train_size:train_size + validation_size]
test_input = data_input[train_size + validation_size:]

train_target = data_target[:train_size]
validation_target = data_target[train_size:train_size + validation_size]
test_target = data_target[train_size + validation_size:]
test_target
```

```python
np.savez('train_data.npz', inputs=train_input, targets=train_target)
np.savez('validation_data.npz', inputs=validation_input, targets=validation_target)
np.savez('test_data.npz', inputs=test_input, targets=test_target)
```

```python
import joblib

scaler_target = StandardScaler()
scaled_target = scaler_target.fit_transform(data[['Kp', 'Ki', 'Kd']])

joblib.dump(scaler_target, 'scaler_target.pkl')
```
